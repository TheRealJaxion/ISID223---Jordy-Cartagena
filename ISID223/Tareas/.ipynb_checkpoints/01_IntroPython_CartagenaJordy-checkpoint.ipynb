## Introducción a Python para Sistemas de Información

### A. Variables y Tipos de Datos
Copia y ejecuta el siguiente código en una celda para registrar los datos de un usuario en un Sistema
de Información Empresarial:

```python
# Declaración de variables (Atributos de una entidad)
nombre_usuario = "Juan Pérez" # String (Texto)
edad = 20 # Integer (Entero)
promedio = 8.5 # Float (Decimal)
es_estudiante = True # Boolean (Lógico)
print(f"Usuario: {nombre_usuario}, Edad: {edad}, Estado: {es_estudiante}")
```

### B. Condicionales (Lógica de Negocio)
Los sistemas aplican reglas. Por ejemplo, determinar si un cliente recibe un descuento:

```python
if edad >= 18:
 mensaje = "Usuario mayor de edad: Acceso total al SI."
else:
 mensaje = "Usuario menor de edad: Acceso restringido."
print(mensaje)
```

### C. Listas y Lazos (Procesamiento de Datos)
Para manejar múltiples registros (como una lista de productos en un inventario):

```python
# Lista de componentes de un SI
componentes_si = ["Hardware", "Software", "Datos", "Redes", "Personas", "Procesos"]
print("Componentes a auditar:")
for item in componentes_si:
 print(f"- {item}")
```

## Introducción a Pandas: Manejo de Información Tabular
Un SI profesional organiza los datos en tablas, que en Python se representan mediante el tipo de dato
DataFrame. Utilizamos la librería pandas para leer estos datos.
### Ejercicio de Lectura de Datos
Copia este bloque para simular la carga de un pequeño dataset de ventas:


```python
import pandas as pd
# Creación de un DataFrame manual (Simulando una base de datos)
datos_empresa = {
 'Producto': ['Laptop', 'Mouse', 'Monitor', 'Teclado'],
 'Precio': [1200, 25, 200, 45],
 'Stock': [15, 100, 20, 50]
}
df = pd.DataFrame(datos_empresa)
# Visualización de los primeros registros
print("Vista previa del Sistema de Inventario:")
print(df.head())
# Operación básica: Calcular el valor total del inventario
total_inventario = (df['Precio'] * df['Stock']).sum()
print(f"\nValor total en activos de hardware: ${total_inventario}")
```
